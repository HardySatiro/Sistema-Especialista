```python
def heuristic_3(c1, c6, c11, canais): # MAXIMO LOCAL
    
    m_c1 = 0
    m_c6 = 0
    m_c11 = 0
    #print(canais)
    for i in range(0,c1):
        if i == 0:
            m_c1 = canais[1][i]
        elif m_c1 < canais[1][i]:
            m_c1 = canais[1][i]
  
    for i in range(0,c6):
        if i == 0:
            m_c6 = canais[6][i]
        elif m_c6 < canais[6][i]:
            m_c6 = canais[6][i]
            
    for i in range(0,c11):
        if i == 0:
            m_c11 = canais[11][i]
        elif m_c11 < canais[11][i]:
            m_c11 = canais[11][i]


    if m_c1 < m_c6 < m_c11:
        return 1
    elif m_c1 < m_c11 < m_c6:
        return 1
    
    elif m_c6 < m_c11 < m_c1:
        return 6
    elif m_c6 < m_c1 < m_c1:
        return 6
    
    elif m_c11 < m_c6 < m_c1:
        return 11
    elif m_c11 < m_c1 < m_c6:
        return 11
  
```
